### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja_1 = pd.read_csv(url)
loja_2 = pd.read_csv(url2)
loja_3 = pd.read_csv(url3)
loja_4 = pd.read_csv(url4)
```

#1. Análise do faturamento


```python
# Faturamento de cada loja

faturamente_total_loja_1 = loja_1['Preço'].sum() # 1534509.12
faturamente_total_loja_2 = loja_2['Preço'].sum() # 1488459.06
faturamente_total_loja_3 = loja_3['Preço'].sum() # 1464025.03
faturamente_total_loja_4 = loja_4['Preço'].sum() # 1384497.58



# Formatando para R$

def formatar(valor):
    return 'R${:,.2f}'.format(valor).replace(',', 'x').replace('.',',').replace('x','.')

valor_formatado_1 = formatar(faturamente_total_loja_1)
print(f"O faturamento da Loja 1 foi de: {valor_formatado_1}")
valor_formatado_2 = formatar(faturamente_total_loja_2)
print(f"\nO faturamento da Loja 2 foi de: {valor_formatado_2}")
valor_formatado_3 = formatar(faturamente_total_loja_3)
print(f"\nO faturamento da Loja 3 foi de: {valor_formatado_3}")
valor_formatado_4 = formatar(faturamente_total_loja_4)
print(f"\nO faturamento da Loja 4 foi de: {valor_formatado_4}")
```

```python
# Dados faturamento

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamento_total = [faturamente_total_loja_1, faturamente_total_loja_2, faturamente_total_loja_3, faturamente_total_loja_4]

plt.figure(figsize=(18, 5))

# Gráficos de barras

cores = plt.cm.tab20c(range(len(lojas)))
plt.subplot(1, 3, 1)
plt.bar(lojas, faturamento_total, color=cores)
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$ milhões)')
plt.title('Faturamento por Loja')


# Grafico de pizza

plt.subplot(1, 3, 2)
plt.pie(faturamento_total, labels=lojas, autopct='%1.1f%%')
plt.title('Distribuição por Loja')

# Gráfico barras horizontal

plt.subplot(1, 3, 3)
plt.barh(lojas, faturamento_total, color=cores)
plt.xlabel('Lojas')
plt.ylabel('Faturamento em milhões de reais')
plt.title('Faturamento por Loja')

plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
categoria_produtos_1 = loja_1['Categoria do Produto'].value_counts()
print(f"Vendas por categoria da Loja 1:\n{categoria_produtos_1.to_string()}")
categoria_produtos_2 = loja_2['Categoria do Produto'].value_counts()
print(f"\nVendas por categoria da Loja 2:\n{categoria_produtos_2.to_string()}")
categoria_produtos_3 = loja_3['Categoria do Produto'].value_counts()
print(f"\nVendas por categoria da Loja 3:\n{categoria_produtos_3.to_string()}")
categoria_produtos_4 = loja_4['Categoria do Produto'].value_counts()
print(f"\nVendas por categoria da Loja 4:\n{categoria_produtos_4.to_string()}")
```

```python
# Gerar gráfico de barra agrupadas

df_categorias = pd.DataFrame({
    'Loja 1': categoria_produtos_1,
    'Loja 2': categoria_produtos_2,
    'Loja 3': categoria_produtos_3,
    'Loja 4': categoria_produtos_4
})

ax = df_categorias.plot(kind='bar', figsize=(12, 6))
ax.set_title('Vendas por Categoria de Produto por Loja')
ax.set_xlabel('Categoria do Produto')
ax.set_ylabel('Número de Vendas')
plt.xticks(rotation=45)
plt.legend(title='Lojas')
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
avaliação_media_1 = loja_1['Avaliação da compra'].mean().round(2)
print(f"A avaliação média da Loja 1 é de: {avaliação_media_1}")
avaliação_media_2 = loja_2['Avaliação da compra'].mean().round(2)
print(f"\nA avaliação média da Loja 2 é de: {avaliação_media_2}")
avaliação_media_3 = loja_3['Avaliação da compra'].mean().round(2)
print(f"\nA avaliação média da Loja 3 é de: {avaliação_media_3}")
avaliação_media_4 = loja_4['Avaliação da compra'].mean().round(2)
print(f"\nA avaliação média da Loja 4 é de: {avaliação_media_4}")
```

```python
notas = [avaliação_media_1, avaliação_media_2, avaliação_media_3, avaliação_media_4]

plt.figure(figsize=(13, 9))

for i, (lojasi, notasi) in enumerate(zip(lojas, notas)):
    plt.annotate(f'({lojasi}, {notasi})', (lojasi, notasi), textcoords="offset points", xytext=(0, 10), ha='center')

plt.title('Gráfico da Média de Avaliações de Compra')
plt.plot(lojas, notas)
plt.grid(True)
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
lista_produtos_1 = loja_1['Produto'].value_counts()
mais_vendido_1 = lista_produtos_1.nlargest(5)
print("Os produtos mais vendidos da Loja 1 foram:")
for produto, quantidade in mais_vendido_1.items():
    print(f"{produto} com: {quantidade} unidades")


menos_vendido_1 = lista_produtos_1.nsmallest(5)
print("\nOs produtos menos vendidos da Loja 1 foram:")
for produto, quantidade in menos_vendido_1.items():
    print(f"{produto} com: {quantidade} unidades")


lista_produtos_2 = loja_2['Produto'].value_counts()
mais_vendido_2 = lista_produtos_2.nlargest(5)
print("\nOs produtos mais vendidos da Loja 2 foram:")
for produto, quantidade in mais_vendido_2.items():
    print(f"{produto} com: {quantidade} unidades")


menos_vendido_2 = lista_produtos_2.nsmallest(5)
print("\nOs produtos menos vendidos da Loja 2 foram:")
for produto, quantidade in menos_vendido_2.items():
    print(f"{produto} com: {quantidade} unidades")


lista_produtos_3 = loja_3['Produto'].value_counts()
mais_vendido_3 = lista_produtos_3.nlargest(5)
print("\nOs produtos mais vendidos da Loja 3 foram:")
for produto, quantidade in mais_vendido_3.items():
    print(f"{produto} com: {quantidade} unidades")


menos_vendido_3 = lista_produtos_3.nsmallest(5)
print("\nOs produtos menos vendidos da Loja 3 foram:")
for produto, quantidade in menos_vendido_3.items():
    print(f"{produto} com: {quantidade} unidades")



lista_produtos_4 = loja_4['Produto'].value_counts()
mais_vendido_4 = lista_produtos_4.nlargest(5)
print("\nOs produtos mais vendidos da Loja 4 foram:")
for produto, quantidade in mais_vendido_4.items():
    print(f"{produto} com: {quantidade} unidades")


menos_vendido_4 = lista_produtos_4.nsmallest(5)
print("\nOs produtos menos vendidos da Loja 4 foram:")
for produto, quantidade in menos_vendido_4.items():
    print(f"{produto} com: {quantidade} unidades")
```

```python
# Criação data frame mais vendidos

df_mais_vendidos = pd.DataFrame({
    'Loja 1': mais_vendido_1,
    'Loja 2': mais_vendido_2,
    'Loja 3': mais_vendido_3,
    'Loja 4': mais_vendido_4
})

# Criação data frame menos vendidos

df_menos_vendidos = pd.DataFrame({
    'Loja 1': menos_vendido_1,
    'Loja 2': menos_vendido_2,
    'Loja 3': menos_vendido_3,
    'Loja 4': menos_vendido_4
})

# Concatenando loja 1
mais_menos_loja_1 = pd.concat([df_mais_vendidos['Loja 1'], df_menos_vendidos['Loja 1']], axis=1)
mais_menos_loja_1.columns = ['Mais Vendidos', 'Menos Vendidos']  # Nomeando as colunas

# Plotando gráfico loja 1
ax1 = mais_menos_loja_1.plot(kind='bar', figsize=(12, 6))
ax1.set_title('Mais e menos vendidos da Loja 1')
ax1.set_xlabel('Produtos')
ax1.set_ylabel('Número de Vendas')
plt.tight_layout()
plt.show()

# Concatenando loja 2
mais_menos_loja_2 = pd.concat([df_mais_vendidos['Loja 2'], df_menos_vendidos['Loja 2']], axis=1)
mais_menos_loja_2.columns = ['Mais Vendidos', 'Menos Vendidos']  # Nomeando as colunas

# Plotando gráfico loja 2
ax2 = mais_menos_loja_2.plot(kind='bar', figsize=(12, 6))
ax2.set_title('Mais e menos vendidos da Loja 2')
ax2.set_xlabel('Produtos')
ax2.set_ylabel('Número de Vendas')
plt.tight_layout()
plt.show()

# Concatenando loja 3
mais_menos_loja_3 = pd.concat([df_mais_vendidos['Loja 3'], df_menos_vendidos['Loja 3']], axis=1)
mais_menos_loja_3.columns = ['Mais Vendidos', 'Menos Vendidos']

# Plotando gráfico loja 3
ax3 = mais_menos_loja_3.plot(kind='bar', figsize=(12, 6))
ax3.set_title('Mais e menos vendidos da Loja 3')
ax3.set_xlabel('Produtos')
ax3.set_ylabel('Número de Vendas')
plt.tight_layout()
plt.show()

# Concatenando loja 4
mais_menos_loja_4 = pd.concat([df_mais_vendidos['Loja 4'], df_menos_vendidos['Loja 4']], axis=1)
mais_menos_loja_4.columns = ['Mais Vendidos', 'Menos Vendidos']

# Plotando gráfico loja 4
ax4 = mais_menos_loja_4.plot(kind='bar', figsize=(12, 6))
ax4.set_title('Mais e menos vendidos da Loja 4')
ax4.set_xlabel('Produtos')
ax4.set_ylabel('Número de Vendas')
plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


```python
frete_1 = loja_1['Frete'].mean().round(2)
print(f"O frete médio gasto pela Loja 1 foi de R${frete_1:.2f}")

frete_2 = loja_2['Frete'].mean().round(2)
print(f"\nO frete médio gasto pela Loja 2 foi de R${frete_2:.2f}")

frete_3 = loja_3['Frete'].mean().round(2)
print(f"\nO frete médio gasto pela Loja 3 foi de R${frete_3:.2f}")

frete_4 = loja_4['Frete'].mean().round(2)
print(f"\nO frete médio gasto pela Loja 4 foi de R${frete_4:.2f}")
```

```python
fretes_total = [frete_1, frete_2, frete_3, frete_4]

plt.figure(figsize=(15, 9))

for i, (lojasi, fretes_totali) in enumerate(zip(lojas, fretes_total)):
    plt.annotate(f'({lojasi}, {fretes_totali})', (lojasi, fretes_totali), textcoords="offset points", xytext=(0, 10), ha='center')

plt.title('Gráfico Gasto com Frete das Lojas')
plt.plot(lojas, fretes_total)
plt.grid(True)
plt.show()
```

#6. Análise de Desempenho Geográfico

```python
lat_lon_loja_1 = pd.concat([loja_1['lat'], loja_1['lon']], axis=1)
lat_lon_loja_1.columns = ['Latitude', 'Longitude']

lat_lon_loja_2 = pd.concat([loja_2['lat'], loja_2['lon']], axis=1)
lat_lon_loja_2.columns = ['Latitude', 'Longitude']

lat_lon_loja_3 = pd.concat([loja_3['lat'], loja_3['lon']], axis=1)
lat_lon_loja_3.columns = ['Latitude', 'Longitude']

lat_lon_loja_4 = pd.concat([loja_4['lat'], loja_4['lon']], axis=1)
lat_lon_loja_4.columns = ['Latitude', 'Longitude']



fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Loja 1
axs[0, 0].scatter(lat_lon_loja_1['Longitude'], lat_lon_loja_1['Latitude'], c='blue', marker='o')
axs[0, 0].set_title('Localização da Loja 1')
axs[0, 0].set_xlabel('Longitude')
axs[0, 0].set_ylabel('Latitude')
axs[0, 0].grid(True)

# Loja 2
axs[0, 1].scatter(lat_lon_loja_2['Longitude'], lat_lon_loja_2['Latitude'], c='green', marker='s')
axs[0, 1].set_title('Localização da Loja 2')
axs[0, 1].set_xlabel('Longitude')
axs[0, 1].set_ylabel('Latitude')
axs[0, 1].grid(True)

# Loja 3
axs[1, 0].scatter(lat_lon_loja_3['Longitude'], lat_lon_loja_3['Latitude'], c='red', marker='^')
axs[1, 0].set_title('Localização da Loja 3')
axs[1, 0].set_xlabel('Longitude')
axs[1, 0].set_ylabel('Latitude')
axs[1, 0].grid(True)

# Loja 4
axs[1, 1].scatter(lat_lon_loja_4['Longitude'], lat_lon_loja_4['Latitude'], c='purple', marker='D')
axs[1, 1].set_title('Localização da Loja 4')
axs[1, 1].set_xlabel('Longitude')
axs[1, 1].set_ylabel('Latitude')
axs[1, 1].grid(True)

plt.tight_layout()  # Ajusta o layout para evitar sobreposição
plt.show()
```

**Pior Avaliação dos Clientes**: Embora a diferença entre as avaliações seja pequena, a Loja 4 obteve a menor média de satisfação.

**Desempenho Geográfico Inferior**: Apesar de uma distribuição geográfica similar entre as lojas, a Loja 4 possui menor presença em regiões com maior densidade de vendas.
